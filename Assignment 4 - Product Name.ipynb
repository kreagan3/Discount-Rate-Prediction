```python
import pandas as pd
import numpy as np

#read dataset
df=pd.read_csv("Jill_August_1_shortened.csv")         
df.head()
```

```python
#change variable types
df.dtypes
#df['Price Type']=df['Price Type'].astype('category')
#df['Price Type']=df['Price Type'].cat.codes
#df['productName']=df['Product Name']
df['priceType']=df['Price Type']
df.head()
```

```python
#drop unnecessary columns
df1=df.drop('Product Name',axis=1,inplace=True)

#separate target from dependent variables
target=df.priceType
inputs=df.drop('priceType',axis=1)
```

```python
#change variable type
size=pd.get_dummies(inputs.productName)
```

```python
inputs=pd.concat([inputs,size],axis=1)
inputs.drop('productName',axis=1,inplace=True)
inputs.head()
```

```python
#split dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(inputs,target,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
```

```python
#train model
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
